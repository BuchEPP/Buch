```python
def x(t,a,b):return b*(a-t)/((a-t)**2+b**2)
def y(t,a,b):return b**2/((a-t)**2+b**2)
```

function of $x$:

```python
simplify(y(t,a,b)**2+x(t,a,b)**2-y(t,a,b))
```
