# Pseudo random numbers

The calculations in this notebook are explained in chapter 1.6 and used in exercises 
1.24, 1.25 and 1.26

```python
def ran0(seed):
    # starting seed 
    m = 2**31-1
    a = 7**5
    c=0
    while True:
        seed = (a * seed + c) % m
        yield seed/(m+1.)
```

```python
# Generate 10 random numbers 
x = ran0(1234567)
for i in range(10):
    print(next(x))
```

## Make simple histogram with random numbers

```python
# Make histogram with N random numbers
# and 10 bins in the range(0,1)
# using function ran0
import numpy as np
import matplotlib.pyplot as plt
N = 100
x = ran0(1234567) # get first random number
lran0 = []
for i in range(N-1):
    lran0.append(next(x))
plt.hist(lran0,10,(0,1)) 
plt.show()
```

## Use random numbers from numpy

```python
# use numpy for random number generation (rng)
# rng = np.random.default_rng(1234567) # use user defined seed
rng = np.random.default_rng() # use random seed
print(rng.random(10)) # generate 10 random numbers
plt.hist(rng.random(100),10,(0,1)) # plot 100 random numbers 
plt.show()
```
